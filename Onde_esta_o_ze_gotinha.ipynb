## Vacinas, Brasil e PNI.

A varíola era a doença mais temida no mundo durante o século XVIII. Após milhões de mortes, graças aos estudos de Edward Jenner, foi desenvolvida a primeira vacina da história.  Milhões de vidas foram poupadas e a varíola erradicada do mundo.

No Brasil, a primeira vacina da varíola chegou no ano de 1804, apenas quatro anos após a marinha britânica.

Nos anos de  1837 e 1846, a vacina da varíola se tornou obrigatória, para crianças e adultos, respectivamente. Contudo, grande era a resistência com a vacinação pela população da época, inúmeros eram os <b>boatos</b> sobre a vacinação, pessoas diziam que quem se vacinou <b>desenvolveria feições bovinas</b>. (https://portal.fiocruz.br/noticia/revolta-da-vacina-2)

Em 1900-1901 foi criado o <b>Instituto Soroterápico do Rio de Janeiro</b> (futura Fiocruz), dirigido por <b>Oswaldo Cruz</b>. Em 1904 Oswaldo Cruz, enviou um projeto para reestruturar e enrijecer a obrigatoriedade da vacinação, sugerindo sanções para aqueles que se recusam a vacinar. O projeto foi aprovado no final daquele ano e serviu como estopim para um movimento conhecido como <b>Revolta da vacina</b>.

A população sem acesso à informação não aceitava a vacinação, que muitas vezes acontecia à força. Além disso parte da população que tinha interesses em depor o presidente Rodrigue Alves, monarquistas, militares, e operários, aproveitaram o momento de revolta para criar um grande movimento de insatisfação popular. Ainda nesse ano de 1904 foi criado a <b>Liga Contra a Vacinação Obrigatória</b>.

Após muita revolta, quase mil prisões, mais de 100 feridos e 30 mortes, o presidente desistiu da vacinação obrigatória, pondo fim à revolta. Após essa turbulência social e política, em 1908, o Rio de Janeiro foi assolado pela maior epidemia de varíola da sua história, e em um movimento contrário ao ocorrido em 1904 a população buscou a vacinação em números recordes.(https://portal.fiocruz.br/noticia/revolta-da-vacina-2)

Após esse episódio, e com o desenvolvimento de novas tecnologias e vacinas, a ideia de se vacinar começou a ser mais aceita na população e novos imunizantes foram sendo adotados no Brasil. 

Em 1927 teve início a vacinação contra a tuberculose com a <b>vacina BCG</b> e em 1942 o país já tinha <b>eliminado a febre amarela urbana</b>. 

Após o êxito em erradicar a varíola do Brasil, no final dos anos 60 <b>(0 último caso de varíola notificado no Brasil foi em 1971)</b>, foi formulado em 1973, por determinação do ministério da saúde o <b>Programa Nacional de Imunizações - PNI</b>. 

O programa foi criado com o objetivo de coordenar as ações de imunizações em todo território nacional. Com a premissa de uma <b>saúde sem excluídos</b>, o PNI tornou-se 'ação de governo caracterizada pela inclusão social, na medida em que assiste todas as pessoas, em todos os recantos do País, sem distinção de qualquer natureza. Seja rico ou pobre, more no litoral ou nos sertões, seja velho ou jovem, o brasileiro sabe que pode contar com vacina de boa qualidade em todos os momentos de sua vida.' https://bvsms.saude.gov.br/bvs/publicacoes/livro_30_anos_pni.pdf
    
O PNI brasileiro é hoje referência mundial citada pela Pan-Americana da Saúde (OPAS), braço da Organização Mundial de Saúde (OMS). E foi responsável por desenvolver e coordenar diversas campanhas bem sucedidas de vacinação, como por exemplo a campanha de vacinação contra a poliomielite e criação do programa de Programa de Autossuficiência Nacional em Imunobiológicos (Pasni), voltado ao suprimento da demanda nacional de vacinas e soros.
    
Um dos focos do PNI é a comunicação, informação e conscientização da população sobre os benefícios da vacinação. Nesse sentido, em 1987 foi apresentado à população um dos personagens brasileiros mais conhecidos, o <b>Zé gotinha</b>. Com o objetivo de informar e tornar a vacinação menos assustadora para crianças, o personagem foi utilizado pela primeira vez na campanha contra pólio. Após a  erradicação da poliomielite no final dos anos 80, o personagem, já incorporado ao imaginário do brasileiro, foi utilizado em diversas campanhas de vacinação e de conscientização.
    
Desde a criação do PNI a vacinação no brasil é um sucesso e um exemplo. Contudo, nos últimos anos o Brasil viu o número de pessoas vacinadas caindo e o reaparecimento de doenças. 
    
Porque o Brasileiro deixou de buscar vacinação? <b>Onde está o zé gotinha?</b>

Embora contraditório, a principal explicação para a queda da vacinação é o sucesso das campanhas. Como muitas doenças estão controladas e/ou eliminadas, muitos pais nunca tiveram preocupação com a imunização. Desse modo, a procura por vacinação tende a diminuir [https://saude.abril.com.br/medicina/por-que-as-pessoas-estao-tomando-menos-vacina/]. 

Ou seja, esse é um momento ideal para promoção de conscientização e para informar a população sobre a importância das vacinas.

Contudo, mesmo diante da necessidade de informar e conscientizar, desde 2018, os investimentos em campanhas de vacinação vem sofrendo suscetíveis cortes. Segundo levantamento feito pelo UOL (https://noticias.uol.com.br/saude/ultimas-noticias/redacao/2020/12/26/governo-bolsonaro-reduz-gastos-campanhas-vacinacao.htm) em 2019 o orçamento de campanhas de vacinação teve um corte de 21% em relação a 2018, e em 2020 outro corte de 24% em relação ao ano anterior. 

Diante desse cenário de ausência de informações, a vacinação vira campo feril para a disseminação de fake news. Hoje, o movimento antivacina já é considerado um das grandes ameaças à saúde mundial. 
    
Nesse panorama, qual a real situação da imunização no Brasil? Quais os riscos corremos? Qual a situação da cobertura vacinal no país, e como isso pode se relacionar com o reaparecimento de doenças já erradicadas?

## O que é Cobertura vacinal?

A cobertura vacinal é o percentual de indivíduos do grupo alvo imunizados com uma vacina específica em um espaço de tempo em um ano específico.  Ou seja, a cobertura vacinal estima o nível de proteção de uma população contra uma doença.

A cobertura vacinal é calculada seguindo a fórmula: 

Número de pessoas com o esquema básico completo / número de pessoas alvo X 100

Segundo o PNI cada agravo tem uma meta de cobertura vacinal para manter as doenças sob controle.



| Vacina | Meta de cobertura vacinal  |
|------|------|
| BCG | 90% |
| HB | 95% |
| Penta | 95% |
| Polio | 95% |
| Febre Amarela | 100% |
| Tríplice | 95% |
| HPV | 80% |
| Hepatite A | 95% |
| Influenza | 80% |


## Como está a cobertura vacinal no Brasil?

Segundo vários levantamentos a cobertura vacinal brasileira está em queda a vários anos, segundo alguns levantamentos a taxa de cobertura vacinal cai a mais de 5 anos no país. (https://agenciabrasil.ebc.com.br/saude/noticia/2020-10/em-queda-ha-5-anos-coberturas-vacinais-preocupam-ministerio-da-saude)

De fato, no ano de 2019 nenhuma vacina atingiu a meta da cobertura vacinal.

Diante desses números, será o Brasil corre risco de descontrole de alguns agravos devido a baixa cobertura vacinal?

## Como a incidência das doenças se relaciona à cobertura vacinal?

Algumas campanhas no Brasil tiveram enorme sucesso no controle de doenças. Dentre as mais notáveis podemos citar, a poliomielite, sarampo, hepatites, tuberculose e febre amarela. Desse modo, vamos analisar a situação vacinal e o número de casos para agravo específico.

```python
vacinas_hepatite = ['Hepatite A', 'Hepatite B']
vacinas_sarampo = ['Sarampo','Tríplice Viral  D1']
Vacinas_outras = ['Poliomielite','Febre Amarela', 'BCG']

cobertura_hepatite = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @vacinas_hepatite').reset_index()
cobertura_sarampo = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @vacinas_sarampo').reset_index()
cobertura_outras = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @Vacinas_outras').reset_index()
```

### Hepatite A e B

A primeira vacina contra a <b>Hepatite A</b> foi aprovada em 1991 na Europa. Logo depois, em 95 foi aprovada nos Estados Unidos. A administração da vacina contra Hepatite A é recomendada no primeiro ano de vida é <b>a duração da imunização gira em torno de 15 anos</b> https://www.who.int/wer/2012/wer8728_29.pdf . No Brasil a vacina contra Hepatite A <b>entrou no calendário do PNI em 2014.</b>

A primeira vacina contra <b>Hepatite B</b> foi aprovada em 1981 nos Estados Unidos. Atualmente é recomendada a <b>aplicação da primeira dose logo no primeiro dia de vida</b>. O vírus da Hepatite B sofreu inúmeras <b>mutações</b> durante os anos, e hoje existe um grande problema relacionado ao <b>escape vacinal</b> e aos tratamentos estabelecidos.  

### Como anda a cobertura vacinal da Hepatite A e B?

```python
plt.style.use('seaborn-whitegrid')

g = sns.catplot(data = cobertura_hepatite.sort_values('Imuno'), x = 'Ano', y = 'Cobertura Vacinal', kind='bar', col='Imuno',  height=5, aspect=2, color ='#7396B8')
  
g.map(plt.axhline, y=95, ls='--', color='gray', linewidth=2)
plt.tight_layout()

plt.text(-25, 97, 'Meta da cobertra vacinal', fontsize=13, color='gray')
    
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Cobertura vacinal Hepatite A e B', size = 20, color ='gray')
g.set_titles(col_template='{col_name}', color ='gray', size = 14)

g.set_axis_labels('Ano', 'Cobertura Vacinal', size=14, color ='gray')
g.set_xticklabels(color ='gray')

plt.show()
```

Após a incorporação da vacina contra Hepatite A no PNI, apenas no ano de 2015 a vacinação atingiu a meta da cobertura vacinal

Já a vacinação contra Hepatite B atingiu a meta de 2004 até 2016. Após esse período a vacinação teve um desempenho abaixo da meta, registrando em 2019 o pior valor da série histórica (52%). 


```python
g = sns.catplot(data = hepatite_a_b_aberta, x = 'Ano', y = 'Incidência', kind='bar', col ='Hepatites', height=5, aspect=2, color ='#BA8F8F')
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Taxa de incidência de Hepatite A e B', size = 20, color ='gray')

g.set_titles(col_template='{col_name}', color ='gray', size = 14)
g.set_axis_labels('Ano', 'Incidência', size=14, color ='gray')
g.set_xticklabels(color ='gray')


plt.show()
```

Quando refletimos esses dados na incidência (ocorrência de novos casos durante um período especificado em uma população) da doença na população podemos observar alguns fatos curiosos. 

O primeiro ponto interessante é observar a grande diminuição da incidência de Hepatite A após a incorporação da vacina no PNI. Mesmo não atingindo o objetivo, a vacinação demonstra ter enorme impacto no controle de novos casos. Como a transmissão da Hepatite A é majoritariamente fecal-oral, e muito relacionada a contatos próximos principalmente entre crianças em creches, possivelmente a vacinação antes do primeiro ano de vida auxiliou na diminuição da incidência da doença. 

Já em relação à hepatite B a análise dos dados é um pouco mais complexa. A hepatite B é uma infecção sexualmente transmissível. Tanto os valores de cobertura vacinal, quanto os valores de incidência da doença se mantiveram constantes nos últimos anos. Recentemente, apenas a partir de 2017 a cobertura vacinal apresentou queda, que não foi refletida no aumento de casos. Contudo, a Hepatite B é marcada por evolução silenciosa, geralmente com diagnóstico após décadas da infecção http://www.aids.gov.br/pt-br/publico-geral/hv/o-que-sao-hepatites/hepatite-b. Quando os sintomas aparecem, muito tempo se passou, desse modo caso não haja testagem contínua da população é impossível ter uma relato epidemiológico acertado. Desse modo, é possível que essa queda na cobertura vacinal tenha reflexos daqui a alguns anos.  

## Sarampo

Diferente das hepatites, o sarampo é uma doença altamente contagiosa. O vírus é transmitido por meio do contato direto ou pelas vias aéreas, basta que a pessoa contaminada respire perto da outra. O sarampo é tão contagioso que uma pessoa infectada pode transmitir a doença para 90% das pessoas próximas, caso as pessoas não estejam imunizadas(https://bvsms.saude.gov.br/ultimas-noticias/3025-sarampo-sintomas-prevencao-causas-complicacoes-e-tratamento). O sarampo é uma doença grave que pode levar a morte.

Em 2016 a doença foi considerada erradicada do território nacional pela OPAS, contudo, atualmente o sarampo voltou a registrar casos de surtos no país e em todo o mundo. Entre os anos de 2019 e 2020 os estados do Pará, Rio de Janeiro, São Paulo, Paraná e Santa Catarina foram os mais atingidos.



### O resurgimento do sarampo no Brasil esttá relacionado a taxa de cobertura Vacinal?

O esquema vacinal do sarampo sofreu ao longo do tempo inúmeras modificações.  Até o início dos anos 2000 a vacina aplicada contra o sarampo era exclusivamente contra esse agravo. Logo no início da década, essa vacina foi substituída pela vacina tríplice, uma de vacina atenuada, contendo vírus vivos “enfraquecidos” do sarampo, da rubéola e da caxumba (https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-triplice-viral-sarampo-caxumba-e-rubeola-scr).

Atualmente, ainda existe a opção da tetra viral, que além da proteção fornecida pela tríplice ainda fornece proteção contra a catapora. Em geral o PNI adota essa vacina tetra viral como a segunda dose do processo de imunização contra o sarampo.https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-tetraviral-scr-v


```python
plt.style.use('seaborn-whitegrid')
g = sns.catplot(data =cobertura_sarampo.sort_values('Imuno'), x = 'Ano', y = 'Cobertura Vacinal', kind='bar', col='Imuno',  height=5, aspect=2, color ='#7396B8')


g.map(plt.axhline, y=95, ls='--', color='gray', linewidth=2)
plt.tight_layout()

plt.text(-10, 97, 'Meta da cobertra vacinal', fontsize=13, color='gray')
        
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Cobertura das Vacinas contra Sarampo', size = 20, color ='gray')
g.set_titles(col_template='{col_name}', color ='gray', size = 14)

g.set_axis_labels('Ano', 'Cobertura Vacinal', size=14, color ='gray')
g.set_xticklabels(color ='gray')


plt.show()
```

A vacinação contra o sarampo de 1997 a 2016 sempre atingiu a meta de cobertura vacinal do PNI. Contudo, desde de 2017 a meta cobertura não é atingida chegando a níveis alarmantes em 2016.

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20,6))
ax = sns.barplot(data = casos_sarampo, x = casos_sarampo.index, y = 'Casos no Brasil', color ='#BA8F8F')

```

Como o sarampo é uma doença altamente contagiosa, de rápida manifestação e o alto número de casos é sentido logo após a vacinação não atingir a sua meta. Nos anos 90 o último grande pico da doença foi em 1997, como esperado, no ano anterior a cobertura vacinal necessária não foi atingida. Atualmente, a doença retornou ao Brasil em 2017, um ano após o insucesso da campanha de vacinação.

Atualmente a situação brasileira em relação ao sarampo é crítica. Curiosamente, o isolamento consequente da pandemia da Covid-19 pode ajudar a manter os surtos de sarampo localizados em lugares específicos, contudo, sem a retomada da vacinação o Brasil corre grandes riscos de ter o sarampo circulando novamente.


```python
plt.style.use('seaborn-whitegrid')
g = sns.catplot(data =sarampo_surto, x = 'Ano', y = 'Cobertura vacinal', kind='bar', col='UF',  color ='#7396B8', col_wrap=3, height=5, aspect=2)
g.map(plt.axhline, y=95, ls='--', color='gray', linewidth=2)
plt.tight_layout()

g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Cobertura das Vacinas contra Sarampo nos estados que registram surtos', size = 20, color ='gray')
g.set_titles(col_template='{col_name}', color ='gray', size = 14)

g.set_axis_labels('Ano', 'Cobertura Vacinal', size=14, color ='gray')
g.set_xticklabels(color ='gray')
plt.show()
```
