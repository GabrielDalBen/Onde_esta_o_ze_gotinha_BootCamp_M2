## Vacinas, Brasil e PNI.

A varíola era a doença mais temida no mundo durante o século XVIII. Após milhões de mortes, graças aos estudos de Edward Jenner, foi desenvolvida a <b>primeira vacina da história</b>.  Milhões de vidas foram poupadas e a varíola erradicada do mundo. No Brasil, a primeira vacina contra a varíola chegou no ano de 1804, apenas quatro anos após a marinha britânica.

Nos anos de  1837 e 1846, a vacina da varíola  tornou-se obrigatória, para crianças e adultos, respectivamente. Contudo, existia grande resistência com a vacinação pela população da época, inúmeros eram os <b>boatos</b> sobre a vacinação, pessoas diziam que quem se vacinou <b>desenvolveria feições bovinas</b> [[1](https://portal.fiocruz.br/noticia/revolta-da-vacina-2)].

Em 1900-1901 foi criado o <b>Instituto Soroterápico do Rio de Janeiro</b> (futura Fiocruz), dirigido por <b>Oswaldo Cruz</b>. Em 1904 Oswaldo Cruz, enviou um projeto para reestruturar e enrijecer a obrigatoriedade da vacinação, sugerindo sanções para aqueles que se recusam a vacinar. O projeto foi aprovado no final daquele ano e serviu como estopim para um movimento conhecido como <b>Revolta da vacina</b>.

A população sem acesso à informação não aceitava a vacinação, que muitas vezes acontecia à força. Além disso parte da população que tinha interesses em depor o presidente Rodrigue Alves, monarquistas, militares, e operários, aproveitaram o momento de revolta para criar um grande movimento de insatisfação popular. Ainda nesse ano de 1904 foi criado a <b>Liga Contra a Vacinação Obrigatória</b>.

Após muita revolta, quase mil prisões, mais de 100 feridos e 30 mortes, o presidente desistiu da vacinação obrigatória, pondo fim à revolta. Após essa turbulência social e política, em 1908, o Rio de Janeiro foi assolado pela maior epidemia de varíola da sua história, e em um movimento contrário ao ocorrido em 1904 a população buscou a vacinação em números recordes [[1](https://portal.fiocruz.br/noticia/revolta-da-vacina-2)].

Após esse episódio, e com o desenvolvimento de novas tecnologias e vacinas, a ideia de se vacinar começou a ser mais aceita na população e novos imunizantes foram sendo adotados no Brasil. Em 1927 teve início a vacinação contra a tuberculose com a <b>vacina BCG</b> e em 1942 o país já tinha <b>eliminado a febre amarela urbana</b>. Após inumeros êxitos em erradicar algumas doenças do Brasil, foi formulado em 1973, por determinação do ministério da saúde o <b>Programa Nacional de Imunizações - PNI</b>. 

O programa foi criado com o objetivo de coordenar as ações de imunizações em todo território nacional. Com a premissa de uma <b>saúde sem excluídos</b>, o PNI tornou-se 'ação de governo caracterizada pela inclusão social, na medida em que assiste todas as pessoas, em todos os recantos do País, sem distinção de qualquer natureza. Seja rico ou pobre, more no litoral ou nos sertões, seja velho ou jovem, o brasileiro sabe que pode contar com vacina de boa qualidade em todos os momentos de sua vida.'[[2](https://bvsms.saude.gov.br/bvs/publicacoes/livro_30_anos_pni.pdf)]
    
O PNI brasileiro é hoje referência mundial citada pela Pan-Americana da Saúde (OPAS), braço da Organização Mundial de Saúde (OMS). E foi responsável por desenvolver e coordenar diversas <b>campanhas bem sucedidas de vacinação</b>, como por exemplo a campanha de vacinação contra a poliomielite e criação do programa de Programa de Autossuficiência Nacional em Imunobiológicos (Pasni), voltado ao suprimento da demanda nacional de vacinas e soros.
    
Um dos focos do PNI é a <b>comunicação, informação e conscientização</b> da população sobre os benefícios da vacinação. Nesse sentido, em 1987 foi apresentado à população um dos personagens brasileiros mais conhecidos, o <b>Zé gotinha</b>. Com o objetivo de informar e tornar a vacinação menos assustadora para crianças, o personagem foi utilizado pela primeira vez na campanha contra pólio. Após a  erradicação da poliomielite no final dos anos 80, o personagem, já incorporado ao imaginário do brasileiro, foi utilizado em diversas campanhas de vacinação e de conscientização. Desde a criação do PNI a vacinação no Brasil é um sucesso e um exemplo. Contudo, nos últimos anos o Brasil viu o número de pessoas vacinadas caindo e o reaparecimento de doenças. 
    
Porque o Brasileiro deixou de buscar vacinação? <b>Onde está o zé gotinha?</b>

Embora contraditório, a principal explicação para a queda da vacinação é o sucesso das campanhas. Como muitas doenças estão controladas e/ou eliminadas, muitos pais nunca tiveram preocupação com a imunização. Desse modo, a procura por vacinação tende a diminuir[[3](https://saude.abril.com.br/medicina/por-que-as-pessoas-estao-tomando-menos-vacina/)]. Ou seja, esse é um <b>momento ideal para promoção de conscientização</b> e para informar a população sobre a importância das vacinas.

Contudo, mesmo diante da necessidade de informar e conscientizar, desde 2018, os investimentos em campanhas de vacinação vem sofrendo suscetíveis cortes. Segundo levantamento feito pelo UOL [[4](https://noticias.uol.com.br/saude/ultimas-noticias/redacao/2020/12/26/governo-bolsonaro-reduz-gastos-campanhas-vacinacao.htm)] em 2019 o orçamento de campanhas de vacinação teve um corte de 21% em relação a 2018, e em 2020 outro corte de 24% em relação ao ano anterior. Diante desse cenário de ausência de informações, a vacinação vira campo fertil para a disseminação de fake news. Hoje, o <b>movimento antivacina</b> já é considerado um das grandes ameaças à saúde mundial. 
    
Nesse panorama, qual a real situação da imunização no Brasil? Quais os riscos corremos? Qual a situação da cobertura vacinal no país, e como isso pode se relacionar com o reaparecimento de doenças já erradicadas?

## O que é Cobertura vacinal?

A cobertura vacinal é o <b>percentual de indivíduos do grupo alvo imunizados com uma vacina específica</b> em um espaço de tempo em um ano específico.  Ou seja, a cobertura vacinal estima o nível de proteção de uma população contra uma doença.

A cobertura vacinal é calculada seguindo a fórmula: 

<b>Número de pessoas com o esquema básico completo</b> / <b>Número de pessoas alvo</b> X 100

Segundo o PNI cada agravo tem uma meta de cobertura vacinal para manter as doenças sob controle.



| Vacina | Meta de cobertura vacinal  |
|------|------|
| BCG | 90% |
| HB | 95% |
| Penta | 95% |
| Polio | 95% |
| Febre Amarela | 100% |
| Tríplice | 95% |
| HPV | 80% |
| Hepatite A | 95% |
| Influenza | 80% |

## Como a incidência das doenças se relaciona à cobertura vacinal?

Algumas campanhas no Brasil tiveram enorme sucesso no controle de doenças. Dentre as mais notáveis podemos citar, a poliomielite, sarampo, hepatites, tuberculose. Desse modo, vamos analisar a situação vacinal e o número de casos para agravo específico.

```python
vacinas_hepatite = ['Hepatite A', 'Hepatite B']
vacinas_sarampo = ['Sarampo','Tríplice Viral  D1']
tuberculose = ['BCG']
polio = ['Poliomielite']

cobertura_hepatite = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @vacinas_hepatite').reset_index()
cobertura_sarampo = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @vacinas_sarampo').reset_index()
cobertura_tuberculose = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @tuberculose').reset_index()
cobertura_polio = cobertura_vacinal_por_ano_imuno_aberta.query('Imuno in @polio').reset_index()
```

### Hepatite A e B

A primeira vacina contra a <b>Hepatite A</b> foi aprovada em 1991 na Europa. Logo depois, em 95 foi aprovada nos Estados Unidos. A administração da vacina contra Hepatite A é recomendada no primeiro ano de vida é <b>a duração da imunização gira em torno de 15 anos</b> [[6](https://www.who.int/wer/2012/wer8728_29.pdf)] . No Brasil a vacina contra Hepatite A <b>entrou no calendário do PNI em 2014.</b>

A primeira vacina contra <b>Hepatite B</b> foi aprovada em 1981 nos Estados Unidos. Atualmente é recomendada a <b>aplicação da primeira dose logo no primeiro dia de vida</b>. O vírus da Hepatite B sofreu inúmeras <b>mutações</b> durante os anos, e hoje existe um grande problema relacionado ao <b>escape vacinal</b> e aos tratamentos estabelecidos.  

#### Como anda a cobertura vacinal da Hepatite A e B?

```python
#selecionando o estilo do gráfico
plt.style.use('seaborn-whitegrid')

#Produindo o gráfico
g = sns.catplot(data = cobertura_hepatite.sort_values('Imuno'), x = 'Ano', y = 'Cobertura Vacinal', kind='bar', col='Imuno', 
                height=5, aspect=2, color ='#7396B8',col_wrap=1)

#Marcando a linha da meta da cobertura vacinal
g.map(plt.axhline, y=95, ls='--', color='#A9A9A9', linewidth=2)

#escrevendo Meta da cobertura vacinal
plt.text(0, 219, 'Meta da cobertra vacinal', fontsize=13, color='#A9A9A9')

#Setando títulos e lables
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Cobertura vacinal Hepatite A e B', size = 20, color ='#4F4F4F')
g.set_titles(col_template='{col_name}', color ='#4F4F4F', size = 14)
g.set_axis_labels('Ano', 'Cobertura Vacinal em %', size=14, color ='#4F4F4F')
g.set_xticklabels(color ='#4F4F4F', size =12, rotation = 25)

#Imprimindo a figura
plt.show()
```

Após a incorporação da vacina contra Hepatite A no PNI, apenas no ano de 2015 a vacinação atingiu a meta da cobertura vacinal

Já a vacinação contra Hepatite B atingiu a meta de 2004 até 2016. Após esse período a vacinação teve um desempenho abaixo da meta, registrando em 2019 o pior valor da série histórica (52%). 


```python
#selecionando o estilo do gráfico
plt.style.use('seaborn-whitegrid')

#Produindo o gráfico
g = sns.catplot(data = hepatite_a_b_aberta, x = 'Ano', y = 'Incidência', kind='bar', 
                col ='Hepatites', height=4, aspect=2, color ='#BA8F8F')

#Setando títulos e lables
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('Taxa de incidência de Hepatite A e B', size = 20, color ='#4F4F4F')
g.set_titles(col_template='{col_name}', color ='#4F4F4F', size = 14)
g.set_axis_labels('Ano', 'Incidência', size=14, color ='#4F4F4F')
g.set_xticklabels(size=12, color ='#4F4F4F')

#imprimindo a figura
plt.show()
```

Quando refletimos esses dados na incidência (ocorrência de novos casos durante um período especificado em uma população) da doença na população podemos observar alguns fatos curiosos. 

O primeiro ponto interessante é observar a grande <b>diminuição da incidência de Hepatite A</b> após a incorporação da vacina no PNI. Mesmo não atingindo o objetivo, a vacinação demonstra ter enorme impacto no controle de novos casos. Como a transmissão da Hepatite A é majoritariamente fecal-oral, e muito relacionada a contatos próximos principalmente entre crianças em creches, possivelmente a vacinação antes do primeiro ano de vida auxiliou na diminuição da incidência da doença. 

Já em relação à hepatite B a análise dos dados é um pouco mais complexa. A <b>hepatite B é uma infecção sexualmente transmissível</b>. Tanto os valores de cobertura vacinal, quanto os valores de incidência da doença se mantiveram constantes nos últimos anos. Recentemente, apenas a partir de 2017 a cobertura vacinal apresentou queda, que não foi refletida no aumento de casos. Contudo, a Hepatite B é marcada por <b>evolução silenciosa, geralmente com diagnóstico após décadas da infecção</b> [[7](http://www.aids.gov.br/pt-br/publico-geral/hv/o-que-sao-hepatites/hepatite-b)]. Quando os sintomas aparecem, muito tempo se passou, desse modo caso não haja testagem contínua da população é impossível ter uma relato epidemiológico acertado. Desse modo, <b>é possível que essa queda na cobertura vacinal tenha reflexos no registro de novos casos daqui a alguns anos</b>.  

### Sarampo

<b>Diferente das hepatites, o sarampo é uma doença altamente contagiosa</b>. O vírus é transmitido por meio do contato direto ou pelas vias aéreas, basta que a pessoa contaminada respire perto da outra. O sarampo é tão contagioso que <b>uma pessoa infectada pode transmitir a doença para 90% das pessoas próximas</b>, caso as pessoas não estejam imunizadas [[8](https://bvsms.saude.gov.br/ultimas-noticias/3025-sarampo-sintomas-prevencao-causas-complicacoes-e-tratamento)]. O sarampo é uma doença grave que pode levar a morte.

Em <b>2016 a doença foi considerada erradicada do território nacional</b> pela OPAS, contudo, atualmente o sarampo voltou a registrar casos de surtos no país e em todo o mundo. Entre os anos de 2019 e 2020 os estados do Pará, Rio de Janeiro, São Paulo, Paraná e Santa Catarina foram os mais atingidos.



#### O resurgimento do sarampo no Brasil está relacionado a taxa de cobertura Vacinal?

O esquema vacinal do sarampo sofreu ao longo do tempo inúmeras modificações.  Até o início dos anos 2000 a vacina aplicada contra o sarampo era exclusivamente contra esse agravo. Logo no início da década, essa vacina foi substituída pela vacina tríplice, uma de vacina atenuada, contendo vírus vivos “enfraquecidos” do sarampo, da rubéola e da caxumba [[9](https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-triplice-viral-sarampo-caxumba-e-rubeola-scr)].

Atualmente, ainda existe a opção da tetra viral, que além da proteção fornecida pela tríplice ainda fornece proteção contra a catapora. Em geral o PNI adota essa vacina tetra viral como a segunda dose do processo de imunização contra o sarampo [[10](https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-tetraviral-scr-v)].


```python
#selecionando o estilo do gráfico
plt.style.use('seaborn-whitegrid')

#Produindo o gráfico
g = sns.catplot(data =cobertura_sarampo.sort_values('Imuno'), x = 'Ano', y = 'Cobertura Vacinal', kind='bar', col='Imuno', 
                height=5, aspect=2, color ='#7396B8',col_wrap=1)

#Marcando a linha da meta da cobertura vacinal
g.map(plt.axhline, y=95, ls='--', color='#A9A9A9', linewidth=2)

#escrevendo Meta da cobertura vacinal
plt.text(20, 227, 'Meta da cobertra vacinal', fontsize=13, color='#A9A9A9')

#Setando títulos e lables
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Cobertura das Vacinas contra Sarampo', size = 20, color ='#4F4F4F')
g.set_titles(col_template='{col_name}', color ='#4F4F4F', size = 14)
g.set_axis_labels('Ano', 'Cobertura Vacinal em %', size=14, color ='#4F4F4F')
g.set_xticklabels(color ='#4F4F4F', size =12, rotation = 25)

#Imprimindo a figura
plt.show()
```

A vacinação contra o sarampo de 1997 a 2016 sempre atingiu a meta de cobertura vacinal do PNI. Contudo, desde de 2017 a meta cobertura não é atingida chegando a níveis alarmantes em 2016.

```python
#selecionando o estilo do gráfico e tamanho da figura
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20,6))

#Produindo o gráfico
ax = sns.barplot(data = casos_sarampo, x = casos_sarampo.index, y = 'Casos no Brasil', color ='#BA8F8F')

#Setando títulos e lables
plt.title('Casos de Sarampo no Brasil de 1990 a 2019', size=20, color ='#4F4F4F', loc='left')
ax.set_ylabel('Total de casos de Sarampo',  size=14,color ='#4F4F4F')
ax.set_xlabel('Ano', size=14,color ='#4F4F4F')

#Imprimindo a figura
plt.show()
```

Como o sarampo é uma <b>doença altamente contagiosa</b>, <b>de rápida manifestação</b> o aumento do número de casos é sentido logo após a vacinação não atingir a sua meta. Nos anos 90 o último grande pico da doença foi em 1997, como esperado, no ano anterior a cobertura vacinal necessária não foi atingida. Atualmente, <b>a doença retornou ao Brasil em 2017, um ano após o insucesso da campanha de vacinação</b>.

Atualmente a <b>situação brasileira em relação ao sarampo é crítica</b>. Curiosamente, o isolamento consequente da pandemia da Covid-19 pode ajudar a manter os surtos de sarampo localizados em lugares específicos, contudo, sem a retomada da vacinação o <b>Brasil corre grandes riscos de ter o sarampo circulando novamente</b>.

### Tuberculose

Diferente de todas as doenças abordadas até aqui, a tuberculose não é causada por um vírus. Causada pela <b>bactéria</b> <i>Mycobacterium tuberculosis</i> (MTB) que entra no corpo do hospedeiro através das vias respiratórias e normalmente se aloja no pulmão. Contudo a bactéria pode se alojar em outras partes do corpo como por exemplo, ossos (tuberculose óssea), gânglios (tuberculose ganglionar), pleura (tuberculose pleural) e corrente sanguínea (tuberculose miliar) causando tipos diferentes de tuberculose, a última a mais grave [[11](https://www.tuasaude.com/tuberculose/)].

A vacina contra a tuberculose é composta pelo <b>bacilo de Calmette-Guérin (BCG)</b> e foi pela primeira vez aplicada no Brasil em 1927.
A BCG é aplicada em dose única antes dos cinco anos e quase sempre deixa uma <b>cicatriz no local da aplicação</b>.

Embora não pareça ter muita eficiência em evitar a tuberculose pulmonar, a vacina BCG é responsável direta pela <b>diminuição de casos graves</b> de tuberculose miliar [[12](https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-bcg)].

### Poliomelite

A campanha que deu origem ao <b>Zé gotinha</b>, é um exemplo das mais bem sucedidas campanhas de vacinação no mundo. A poliomielite, ou popularmente conhecida como <b>paralisia infantil</b>, é uma doença que desde o início do século XX já era registrada no Brasil. Contudo, apenas na década de 1950, quando os surtos atingiram grandes cidades, as autoridades sanitárias brasileiras começaram a se preocupar com o controle da doença [[13](https://www.scielo.br/j/hcsm/a/9tFSfwSZjFX6NpSvxq9NZws/?lang=pt)]

A poliomielite na maioria das pessoas não produz nenhum sintoma, sendo um infecção silenciosa. Em uma pequena parcela ocorre a chamada forma subclínica que tem quadro semelhante a outras viroses. Infelizmente, em cerca de 1 a cada 200 casos ocorre a forma paralítica que deixa sequelas (paralisia dos membros inferiores) permanentes e pode levar à morte. 

O vírus da pólio (poliovírus) é transmitido de forma similar ao vírus da Hepatite A, contato fecal-oral, ou seja, entra no corpo através da boca por meio de alimentos, água e objetos contaminados com fezes de doentes ou portadores do vírus,  estando altamente associado a baixo acesso a saneamento básico. Além disso, a poliomielite também pode ser transmitida de pessoas para pessoas através de gotículas na fala, tosse ou espirro. 

Ainda nos anos de 1950 a primeira vacina contra a pólio foi desenvolvida. Conhecida como Salk (VOP), essa vacina  era produzida com base em <b>vírus atenuados</b>. Desse modo, uma pequena quantidade do <b>vírus atenuado era liberada nas fezes</b>, assim o vírus atenuado competia na natureza com o poliovírus, acabando por diminuir sua circulação e consequentemente ajudar na imunização.
Atualmente duas vacinas contra a pólio estão disponíveis no PNI, a VOP e a VIP.

No Brasil o último caso de pólio foi registrado em <b>1989</b>. No Mundo, em 2018 foram registrados 33 casos no Afeganistão e Paquistão, ou seja, <b>o poliovírus ainda está em circulação</b> endêmica como na maioria dos casos a pólio é uma infecção silenciosa o vírus pode entrar no país sem ser notado, por isso a vacinação é de extrema importância para evitar sua volta [[14](https://saude.abril.com.br/blog/com-a-palavra/estamos-mesmo-livres-da-poliomielite/)]

Diante desse cenário, como anda a cobertura vacinal no Brasil contra a paralisia infantil.


#### Situação vacinal da polio no Brasil.

De todas as doenças analisadas até aqui a Pólio é o <b>melhor exemplo de como o investimento na vacinação e na conscientização</b> podem ser ferramentas indispensáveis no combate a epidemias. 

O Brasil, desde os anos de 1980 é exemplo mundial no combate a paralisia infantil, e é indispensável que essa luta não seja deixada de lado. Desse modo, analisar a cobertura vacinal pode nos fornecer indicativos da situação atual do país no <b>combate à poliomielite</b>.


```python
#selecionando o estilo do gráfico e tamanho da figura
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20,6))

#Produindo o gráfico
ax = sns.barplot(data = cobertura_polio, x = 'Ano', y = 'Cobertura Vacinal', color ='#7396B8')

#Marcando a linha da meta da cobertura vacinal
plt.axhline(y=90, ls='--', color='#A9A9A9', linewidth=2)

#Setando títulos e lables
plt.title('Cobertura vacinal contra a Poliomelite de 1994 a 2019', size=20, color ='#4F4F4F', loc='left')
ax.set_ylabel('Cobertura vacinal em %',  size=14,color ='#4F4F4F')
ax.set_xlabel('Ano', size=14,color ='#4F4F4F')

#Imprimindo a figura
plt.show()
```

Desde o final dos anos 1990 o Brasil atingia as metas de cobertura vacinal contra a pólio. Contudo, <b>desde o ano de 2016 as campanhas de vacinação não atingem seu objetivo</b> sendo que em 2019 a cobertura vacinal foi menor do que a observada no ano de 1994. 

Assim, o Brasil está em <b>risco de ser atingido novamente por uma nova epidemia de Pólio</b>. O não comprimento dos objetivos vacinais bota em cheque todos os esforços realizados desde os anos 1950 com o objetivo de erradicar essa doença.


## Abandono Vacinal

A taxa de abandono é o <b>cálculo do percentual de vacinados que iniciaram o esquema vacinal e não finalizaram</b>. ou seja, é a taxa de pessoas que tomaram a primeira dose de alguma vacina e não voltaram para tomar as outras doses.

Alguns especialistas relacionam a taxa de abandono a diversos fatores como a <b>desigualdade social, a falta de estratégias e a desinformação da </b>população [[15](https://brasil.elpais.com/brasil/2021-04-20/de-fake-news-a-desigualdade-o-que-leva-brasileiros-a-nao-voltarem-para-tomar-a-segunda-dose-da-vacina.html)].

Assim, o abandono vacinal além de representar <b>aumento de risco de propagação de doença</b>, também representa um <b>prejuízo financeiro</b>, já que doses de vacinas são perdidas em imunizações incompletas.


```python
#selecionando o estilo do gráfico e tamanho da figura
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(20,6))

ax = sns.barplot(data = taxa_abandono_vacinal_por_ano, x = 'Ano', y = 'Cobertura Vacinal', color ='#DFBE69')

#Setando títulos e lables
plt.title('Taxa de abandono vacinal de 201 a 2021', size=20, color ='#4F4F4F', loc='left')
ax.set_ylabel('Taxa de abandono vacinal em %',  size=14,color ='#4F4F4F')
ax.set_xlabel('Ano', size=14,color ='#4F4F4F')

#Imprimindo a figura
plt.show()
```

O PNI considera valores abaixo de <b>5% de abandono com baixo</b>, entre <b>5% e 10% médio</b> e <b>acima de 10% elevado</b>. Desse modo, desde 2015 o abandono vacinal no brasil é <b>elevado</b>.

O ano de 2019 apresentou o maior percentual de abandono vacinal registrado desde 2015. Mais de 20% das pessoas não completaram o processo de vacinação de pelo menos um imunizante. 2020 também apresentou uma taxa elevada de abandono vacinal e 2021 já apresenta valores de abandono maiores que o ano de 2016 inteiro.

É fato que a pandemia de COVID-19 desempenhou um papel importante no abandono vacinal nos anos de 2020 e 2021, já que a recomendação vigente é de distanciamento social. Contudo, os <b>valores altos observados antes da epidemia evidenciam que o problema é muito maior</b>.

## Conclusão

Diante de todos esses dados, podemos concluir que a situação vacinal do Brasil é crítica, por anos o país não vem atingindo os valores ideais de cobertura vacinal. De fato, atualmente o país apresenta os piores valores já registrados. A pandemia de COVID-19 agravou a situação, contudo esse problema é anterior à pandemia.

Doenças como o Sarampo, antes erradicadas, já voltaram a ocorrer no Brasil logo após a queda na vacinação. Além disso, corremos o risco de termos novamente a Poliomielite circulando no território nacional.

Desse modo, é urgente a retomada do investimento e de políticas públicas que foquem na conscientização e na informação da população sobre a vacinação. Adicionalmente, é necessário garantir o acesso fácil às vacinas.
